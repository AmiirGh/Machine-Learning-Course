```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-100*x))

# Generate x values
x_values = np.linspace(-7, 7, 100)

# Calculate corresponding y values using the sigmoid function
y_values = sigmoid(x_values)

# Plot the sigmoid function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Sigmoid Function', color='blue')
plt.title('Sigmoid Function for w=100')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.axhline(0.5, color='red', linestyle='--', linewidth=1, label='Decision Boundary')
plt.legend()
plt.grid(True)
plt.show()
```
