```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Generating the dataset
N = 1000
np.random.seed(1)
X = np.concatenate((np.random.normal(0, 1, int(0.3 * N)), np.random.normal(5, 1, int(0.7 * N))))[:, np.newaxis]

plt.hist(X, bins=30, density=True, color='red', alpha=0.5)
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Histogram of X')
```

```python
#def gaussian_kernel(x, sigma):
#    return np.exp(-0.5 * (x / sigma) ** 2)

def parzen_window_estimation(x, X, h):
    return np.mean(norm.pdf((x - X) / h) / h)
```

```python
# Define the range of x values for estimation
x_range = np.linspace(-5, 10, 500)

# Set the bandwidth for the Gaussian kernel
h = 0.5

# Perform estimation for each x value
estimated_density = np.array([parzen_window_estimation(x, X, h) for x in x_range])

# Plot the result
plt.plot(x_range, estimated_density, label='Estimated Density')
plt.hist(X, bins=30, density=True, alpha=0.5, color='red', label='Histogram of X')
#plt.hist(X, bins=30, density=True)
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Estimation using Parzen Window (Gaussian Kernel)')
plt.legend()

plt.show()
```
